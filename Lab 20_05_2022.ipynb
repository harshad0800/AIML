```python
#line plot
from matplotlib import pyplot as plt
#x-axis values
x = [5,2,9,4,7]
#y axis-values
y = [10,5,8,4,2]
#function to plot
plt.plot(x,y)
plt.show()
```

```python
#bar plot
from matplotlib import pyplot as plt
#x-axis values
x = [5,2,9,4,7]
#y axis-values
y = [10,5,8,4,2]
#function to plot
plt.bar(x,y)
plt.show()
```

```python
#bar plot
from matplotlib import pyplot as plt
#x-axis values
x = [5,2,9,4,7]
#function to plot
plt.hist(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
#x-axis values
x = [5,2,9,4,7]
#y-axis values
y = [10,5,8,4,2]

plt.scatter(x,y)
plt.title("Sports Data")
plt.xlabel("Avg Value")
plt.ylabel("Calorie Burn")
plt.colorbar()

#function to show the plot
plt.show()
```
