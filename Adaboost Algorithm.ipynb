```python
#Data Pre-processing Step  
# importing libraries  
import numpy as nm    
import pandas as pd  
```

```python
#importing datasets  
data = pd.read_csv("income.csv") 
data.head()
```

```python
#Separating predictors and response
x=data.iloc[:,0:6].values
y =data.iloc[:,6].values
```

```python
#Splitting the dataset into training and test samples
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2,random_state=1)
```

```python
#Initializing Adaboost classifier and fitting the training data
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier(n_estimators=100, base_estimator= None,learning_rate=1)
#train Adaboost classifier
adaboost.fit(x_train,y_train)
```

```python
#predict the response for test dataset
y_pred = adaboost.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy Score for Adaboost Algo',accuracy_score(y_test, y_pred),'\n')
```
