```python
people = {1:{'name':"John",'age':25,'gender':'Male'},
          2:{'name':"Maria",'age':24,'gender':'Female'}}

print(people[1]['name'])
print(people[1]['age'])
print(people[1]['gender'])

people = {1:{'name':"John",'age':25,'gender':'Male'},
          2:{'name':"Maria",'age':24,'gender':'Female'}}

people[3] = {}
people[3]['name'] = 'Luna'
people[3]['age'] = '24'
people[3]['gender'] = 'Male'
people[3]['married']= 'No'
print(people[3])

del people[3]['married']
print(people)
```
